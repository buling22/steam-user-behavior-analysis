# Steam用户行为分析

## 项目背景
基于Kaggle Steam-200k数据集，对约12,000名Steam用户的
购买与游玩行为进行分析，挖掘用户分层特征与平台运营规律。
数据采集于2017 核## 核心结论

**1. 游戏热度分析**
Dota2以4,841人购买量和98万小时总游戏时长双榜第一，
是2016-2017年Steam最具统治力的游戏，购买量是第二名
Team Fortress 2（2,323人）的2倍以上。平均游戏时长榜中
Football Manager系列包揽Top4（2014版392h、2012版390.5h、
2015版315.4h、2013版310.7h），粘性远超其他游戏但购买量
不突出，说明该系列受众圈小但极度忠诚。购买量高但游玩时长
极低的游戏（如Deathmatch Classic 0.7h、Half-Life 
Deathmatch Source 0.8h）存在明显体验落差，可玩性存疑。

**2. 用户购买行为**
45.9%的用户仅购买1款游戏（5,693人），近半数用户属于
尝鲜型，未形成深度使用习惯。购买20款以上的用户仅占9.8%
（1,213人），但属于平台最高价值用户群体。购买量最高的
单一用户达1,075款，属于对所有品类高度接受的重度收藏型玩家。
用户「最爱游戏」分布中，Dota2以3,971人独占第一，
是第二名Team Fortress 2（1,033人）的近4倍，
射击类游戏（CS系列合计约1,170人）整体忠诚度也极高。

**3. 用户分层分析**
轻度用户（总游玩<10h）占比最高达42%（4,763人）；
超级用户（>500h）占15.5%（1,757人），人均购买46.6款，
是轻度用户（2款）的23倍，体现典型二八效应——
少数核心用户贡献了平台绝大部分消费行为。

**4. 沉默用户分析**
12,393名购买用户中有1,043人（8.4%）从未启动任何游戏，
反映了Steam平台特有的「折扣囤游戏」现象，
促销驱动购买但实际游玩意愿不足。

## 业务建议

- **激活沉默用户：** 针对购买后7天未启动的用户推送游戏
  新手引导或好友动态通知，将8.4%的沉默用户转化为活跃用户，
  直接提升DAU。

- **引导轻度用户升层：** 45.9%用户仅购买1款游戏，
  可在用户首款游戏游玩达5小时后推荐同品类低价游戏或捆绑包，
  将轻度用户逐步转化为中度用户，提升平台ARPU值。
